```python
arr = [1,2,3,4,5]
k = 3
# create a sliding widnow
# keep track of the max value
# return the values in a sliding window in clock wise .
# k = 3-> [[1,2,3],[2,3,4],[3,4,5],[4,5,1],[5,1,2]]
# left and right pointer which points to the 0 and k-1

left = 0
right = k-1
end = len(arr) - 1

# 4-> 0
# until left reaches the end which n-1
# once the right reaches the end or it exceed end then set that to 0
result = []
while left <= end:
    if right > end:
        right = 0
    if right < left:
        result.append(arr[left:] + arr[:right+1])
    else:
        result.append(arr[left:right+1])
    left += 1
    right += 1
print(result)

```

```python
arr = [10, 2, 3, 4, 5]
k = 3
n = len(arr)

# Step 1: Initialize sum with first k elements (wrapping included)
window_sum = sum(arr[i % n] for i in range(k))
max_sum = window_sum

# Step 2: Slide window through the array circularly
for i in range(1, n):
    # Remove the outgoing element and add the incoming element
    out_idx = (i - 1) % n
    in_idx = (i + k - 1) % n
    window_sum = window_sum - arr[out_idx] + arr[in_idx]
    max_sum = max(max_sum, window_sum)

print(max_sum)  # Output: 19
```

```python
from bintrees import AVLTree
from typing import List

class Solution:
    def medianSlidingWindow(self, nums: List[int], k: int) -> List[float]:
        # Initialize two AVL trees
        lower = AVLTree()  # Lower half (smaller elements)
        upper = AVLTree()  # Upper half (larger elements)
        result = []
        
        # Initialize first window
        initial_array = sorted(nums[:k])
        for i in range(k):
            if i < (k + 1) // 2:  # Lower half (including middle for odd k)
                lower.insert(initial_array[i], initial_array[i])
            else:  # Upper half
                upper.insert(initial_array[i], initial_array[i])
        
        # Compute median for first window
        if k % 2 == 1:
            median = lower.max_key()
        else:
            median = (lower.max_key() + upper.min_key()) / 2
        result.append(float(median))
        
        # Process sliding windows
        for i in range(k, len(nums)):
            # Remove element nums[i-k]
            if lower.get(nums[i - k]) is not None:  # Check if in lower
                lower.remove(nums[i - k])
            elif upper.get(nums[i - k]) is not None:  # Check if in upper
                upper.remove(nums[i - k])
            # Note: If element not found in either, it's an error in logic, but test cases ensure it's present
            
            # Insert new element
            if not lower or nums[i] <= lower.max_key():
                lower.insert(nums[i], nums[i])
            else:
                upper.insert(nums[i], nums[i])
            
            # Balance the AVL trees
            while len(lower) > len(upper) + 1:
                val = lower.max_key()
                upper.insert(val, val)
                lower.remove(val)
            while len(lower) < len(upper):
                val = upper.min_key()
                lower.insert(val, val)
                upper.remove(val)
            
            # Compute median
            if k % 2 == 1:
                median = lower.max_key()
            else:
                median = (lower.max_key() + upper.min_key()) / 2
            result.append(float(median))
        
        return result
```
